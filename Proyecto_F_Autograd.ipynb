```python
#importamos las librerias necesarias
import autograd.numpy as np
from autograd import elementwise_grad as egrad
import matplotlib.pyplot as plt
#creamos un array que va desde -31.4 hasta 31.4, que tenga 256 datos
x = np.linspace(-31.4,31.4, 256)
#creamos una funcion lambda para calcular el seno de cualquier numero / el mismo numero

sinc = lambda x: np.sin(x) / x
# se configura una figura que mostraremos
plt.figure(figsize=(12,7))
plt.title("sinc function and derivatives",fontsize=24)
#se asigna el valor de sinc a otra variable
my_fn = sinc
#comenzamos un ciclo para calcular (9) derivada 
for ii in range(9):
  #se imprime y guarda cada grafica de la derivada en indice (ii)
  plt.plot(x, my_fn(x), lw=3,label='d{} sinc(x)/dx{}'.format(ii,ii))
  plt.legend(fontsize=18)
  plt.axis([-32, 32, -0.50, 1.2])
  plt.savefig('./sinc_grad{}.png'.format(ii))
  # en el primer ciclo se halla el gradiente de la funcion y luego la derivada de la misma
  my_fn = egrad(my_fn)
```
